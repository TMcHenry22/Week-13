```python
import numpy as np
# your solution here
class Particle(object):
#Q1 
    def __init__(self, mass, x,y,z, vx,vy,vz):
        self.mass=mass
        self.x=x
        self.y=y
        self.z=z
        self.vx=vx
        self.vy=vy
        self.vz=vz
#Q2
    def kineticEnergy(self):
        return (0.5*self.mass*(np.sqrt((self.vx)**2+(self.vy)**2+(self.vz)**2))**2)
#Q3        
        
    def seppt(self,other):
        out=0
        out= np.sqrt((self.x-other.x)**2 + (self.y - other.y)**2 + (self.z - other.z)**2)
        if(out<0.25):
            print( "position interaction")
        return out
    
     
 #   def sepvt(self,other):
 #       return np.sqrt((self.vx-other.vx)**2 + (self.vy - other.vy)**2 + (self.vz - other.vz)**2)
    
#Q4
    def newpt(self,dt):
        self.x= self.x+(self.vx*dt)
        self.y= self.y+(self.vy*dt)
        self.z=self.z+(self.vz*dt)
        
```

```python
P12 = Particle(5,2,2,2,1,1,1)
P21= Particle(5,1,1,1,2,2,2)
print(P12.x)
print(P21.x)
print(P12.kineticEnergy())
print(P12.seppt(P21))
```
